```python
import re
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("data/Algerian_forest_fires_dataset_UPDATE.csv")
```

```python
data["Bejaia Region Dataset "] = data["Bejaia Region Dataset "].str.replace(r'\s', '', regex=True)
data = data.reset_index()
```

```python
column_names = data.iloc[0, :].str.replace(r'\s', '', regex=True).values
header_index = data[data["Bejaia Region Dataset "] == "Classes"].index
```

```python
first_half = data.iloc[header_index[0] + 1:header_index[1]-1, :]
second_half = data.iloc[header_index[1]+1:, :]
```

```python
col_name_dict = dict(zip(first_half.columns, column_names))
first_half = first_half.rename(col_name_dict, axis=1)
second_half = second_half.rename(col_name_dict, axis=1)
```

```python
first_half["Region"] = "Bejaia"
second_half["Region"] = "Sidi-Bel Abbes"
cleaned = pd.concat([first_half, second_half], axis=0)
cleaned = cleaned.dropna()
print(f"Cleaned data have shape {cleaned.shape}")
cleaned.head()
```

```python
cleaned.to_csv("data/Algerian_forest_fires_dataset_CLEANED.csv")
```
